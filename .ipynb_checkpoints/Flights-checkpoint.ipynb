```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv('query_result_2021-06-14T15_59_05.825731Z.csv')
```

```python
df['flight_time'] = df['flight_time'].str.replace('0 years 0 mons 0 days 0 hours','')
df['flight_time'] = df['flight_time'].str.replace('0 years 0 mons 0 days','')
df['flight_time'] = df['flight_time'].str.replace(' 0.00 secs','')
df['Fuel_price_t'] = 45000 # цена в Анапе на декабрь 2017г
df['Fuel_consumption_t/h'] = np.where(df.model == 'Sukhoi Superjet-100', 1.7, 1.857)
df['time_factor'] = np.where(df.model == 'Sukhoi Superjet-100', 0.83, 1.66) 
df['fuel_cost'] =df['Fuel_consumption_t/h'] *df['Fuel_price_t'] *df['time_factor']
```
